```python
#imports modules

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```python
#given initial state (psi0), epsilon, delta, and time list (tlist)
#Returns expectation value arrays for each Pauli operator expectation value at the corresponding time in tlist

def Simulate(psi0,epsilon,delta,tlist):
    H = epsilon/2 * sigmaz() + delta/2 * sigmax()
    e_ops = [sigmax(), sigmay(), sigmaz()]
    result = sesolve(H,psi0,tlist,e_ops)
    return result
```

```python
#parameters

psi0=basis(2,0)
epsilon=6
delta=4
time=3
tlist=np.linspace(0,time,200)

#runs simulation and assigns the arrays for sigmax, sigmay, sigmaz operators' expectation values to sx, sy, sz variables respectively

sx,sy,sz = Simulate(psi0,epsilon,delta,tlist).expect
```

```python
#plots expectation value vs time for sigmax (red), sigmay (green), sigmaz (blue)

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(tlist, np.real(sx), 'r')
ax.plot(tlist, np.real(sy), 'g')
ax.plot(tlist, np.real(sz), 'b')
ax.axhline(color='k',lw=1)
ax.axvline(color='k',lw=1)
ax.legend(("sigma x","sigma y","sigma z"))
ax.set_xlabel('Time')
ax.set_ylabel('Expectation Values');
```
